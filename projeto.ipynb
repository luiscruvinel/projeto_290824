```python


def reverso(string):
    '''
    Recebe uma string e retorna a string reversa
    Exemplo:
        input: "ATCG"
        returns: "CGTA"
    '''

    string_reversa = string[::-1]
    return string_reversa

def reverso_complementar(string):
    '''
    Recebe uma string e retorna seu reverso complementar

    Exemplo:
        input: "ATCG"
        returns: "CGAT"
    '''
    string_reversa = reverso(string)

    decode = {
        'A':'T',
        'T':'A',
        'C':'G',
        'G':'C'
    }

    lista_reversa_complementar = []

    for nuc in string_reversa:
        lista_reversa_complementar.append(decode[nuc])
    
    string_reversa_complementar = ''.join(lista_reversa_complementar)
    return string_reversa_complementar

def transcricao(string):
    '''
    Recebe uma string e retorna a transcrição em RNA

    Exemplo:
        input: "ATCG"
        returns: "UAGC"
    '''
    decode = {
        'A':'U',
        'T':'A',
        'C':'G',
        'G':'C'
    }

    list_RNA = []

    for nuc in string:
        list_RNA.append(decode[nuc])
    
    string_transcrita = ''.join(list_RNA)
    return string_transcrita

def traducao(string):
    '''
    Recebe uma string e retorna a traducao da proteina

    Exemplo:
        input: "código genético"
        returns: "aminoácido"
    '''
    decodificar = {"UUU": "PHE", "UUC": "PHE", "UUA": "LEU", "UUG":"LEU", "CUU":"LEU", "CUC":"LEU", "CUA": "CUG", "AUC":"ISO", "AUU":"ISO", "AUA":"ISO", "AUG":"MET", "GUU":"VAL",
          "GUC":"VAL", "GUA":"VAL", "GUG":"VAL", "UCU":"SER", "UCC":"SER", "UCA":"SER", "UCG":"SER","CCU":"PRO", "CCC":"PRO", "CCA":"PRO", "CCG":"PRO","ACU":"THR","ACC":"THR",
          "ACA": "THR", "ACG": "THR", "GCU":"ALA", "GCC":"ALA", "GCA":"ALA", "GCG":"ALA","UAU":"TYR","UAC":"TYR", "UAA":"STOP","UAG":"STOP", "CAU":"HIS", "CAC":"HIS", "CAA":"GLN",
          "CAG": "GLN", "AAU":"ASN", "AAC":"ASN", "AAA":"LYS", "AAG":"LYS", "GAU": "ASP", "GAC": "ASP", "GAA": "GLU", "GAG": "GLU", "UGU":"CYS", "UCG":"CYS","UGA": "STOP",
          "CGU":"ARG", "CGC":"ARG", "CGA":"ARG", "CGG":"ARG", "AGU":"SER", "AGC":"SER", "AGA":"ARG", "AGG":"ARG", "GGU":"GLY", "GGC":"GLY", "GGA":"GLY", "GGG":"GLY"}


    codons = [string[i:i+3] for i in range(0, len(string), 3)]
    aminoacidos = []

    for i in codons:
      aminoacidos.append(decodificar[i])

    return aminoacidos

    
if __name__ == '__main__':
    string_teste = 'ATCGATCGATCG'
    print(reverso(string_teste))
    print(reverso_complementar(string_teste))
    print(transcricao(string_teste))
    print(traducao('UAGCUAGCUAGC'))
```
